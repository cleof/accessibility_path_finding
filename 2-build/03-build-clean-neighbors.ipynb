```python
nodes = pd.read_csv(mst_path + 'clean_node.csv')
nodes.head()
```

```python
edges = pd.read_csv(mst_path + 'clean_edge.csv')
edges.head()
```

```python
y,x = nodes.loc[0, 'y'],nodes.loc[0, 'x']
ends = edges[(edges['y_start'] == y) & (edges['x_start'] == x)][['y_end', 'x_end']].values
starts = edges[(edges['y_end'] == y) & (edges['x_end'] == x)][['y_start', 'x_start']].values
print(ends)
print(starts)
```

```python
edges[(edges['y_start'] == y)]
```

```python
neighbors = {}
for idx in nodes.index:
    y,x = nodes.loc[idx, 'y'], nodes.loc[idx, 'x']
    neighbors[idx] = find_neighbors(nodes, edges, y, x)
```

```python
neighbors
```

```python
with open(mst_path + 'neighbors.pickle', 'wb') as f:
    pickle.dump(neighbors, f, protocol=pickle.HIGHEST_PROTOCOL)

with open(mst_path + 'neighbors.pickle', 'rb') as f:
    neighbors_p = pickle.load(f)
    
assert neighbors_p == neighbors
```
