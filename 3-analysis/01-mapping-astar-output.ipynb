```python
fig = visualize_map(nodes_df)
plt.scatter(nodes_df.loc[start, 'x'], nodes_df.loc[start, 'y'], c = 'r', label = 'start', marker = '*')
plt.scatter(nodes_df.loc[end, 'x'], nodes_df.loc[end, 'y'], c = 'g', label = 'end', marker = '*')
plt.legend()
for i in range(len(all_paths)):
    plt.scatter(nodes_df.loc[all_paths[i], 'x'], nodes_df.loc[all_paths[i], 'y'], c = colors[i], s = 1)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.savefig(output_path + str(start) + "_" + str(end) + "_whole")
```

```python
fig = visualize_found_path(nodes_df, all_paths, COSTS)
plt.scatter(nodes_df.loc[start, 'x'], nodes_df.loc[start, 'y'], c = 'r', label = '', marker = '*')
plt.scatter(nodes_df.loc[end, 'x'], nodes_df.loc[end, 'y'], c = 'g', label = '', marker = '*')
plt.legend()
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.savefig(output_path + str(start) + "_" + str(end) + "_path")
```

```python
for i in range(len(all_paths)):
    sample = all_paths[i][::5]
    sns.barplot(sample, nodes_df.loc[sample, 'elevation'], palette="Blues")
    plt.xticks(rotation = 90)
    plt.xlabel("node indices")
    plt.title(str(start) + " to " + str(end)+" path_"+str(i))
    plt.savefig(output_path + str(start) + "_" + str(end) + "path_"+str(i))
    plt.show()
```
